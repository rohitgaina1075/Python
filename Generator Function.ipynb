```python
def test_fib(n):   # kis tera ka function ko generator function bolta hai
    a,b = 0,1
    for i in range(n):
        yield a         #yoeld function data ko thro karna ka kam ker raha hai 
        a,b = b, a+b
```

```python
test_fib(10)
```

```python
for i in test_fib(11):
    print(i)
```

```python
def test_fib1():   
    a,b = 0,1
    while True:
        yield a         
        a,b = b, a+b
```

```python
fib = test_fib1()
```

```python
def count_text(n):
    count = 1
    while count < n:
        yield count
        count = count +1
```

```python
c = count_text(5)
```

```python
for i in c :
    print(i)
```
